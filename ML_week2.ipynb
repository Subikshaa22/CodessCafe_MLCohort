## DECISION TREE

```python
import pandas as pd
import io
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['MBA.csv']))
df
```

```python
df.fillna({'admission' : 'Not Admit'}, inplace=True)
df.fillna({'race' : 'Other'}, inplace = True)
```

```python
inputs = df.drop('admission',axis='columns')
target = df['admission']
```

```python
target = target.map({'Not Admit': 0, 'Admit': 2, 'Waitlist': 1})

target
```

```python
from sklearn.preprocessing import LabelEncoder
le_gender = LabelEncoder()
le_international = LabelEncoder()
le_major = LabelEncoder()
le_race = LabelEncoder()
le_work_industry = LabelEncoder()
```

```python
inputs['gender_n'] = le_gender.fit_transform(inputs['gender'])
inputs['international_n'] = le_international.fit_transform(inputs['international'])
inputs['major_n'] = le_major.fit_transform(inputs['major'])
inputs['race_n'] = le_race.fit_transform(inputs['race'])
inputs['work_industry_n'] = le_work_industry.fit_transform(inputs['work_industry'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['application_id','major','work_industry','international','gender','race'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[3.47,690.0,6.0,1,0,2,1,13]])
```

## K NEAREST NEIGHBOURS


```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['bank.csv']), sep = ';', header = 0)
df
```

```python
inputs = df.drop('y',axis='columns')
target = df['y']
```

```python
target = target.map({'no': 0, 'yes': 1})
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_job = LabelEncoder()
le_marital = LabelEncoder()
le_education = LabelEncoder()
le_default = LabelEncoder()
le_housing = LabelEncoder()
le_loan = LabelEncoder()
le_contact = LabelEncoder()
le_month = LabelEncoder()
le_poutcome = LabelEncoder()
```

```python
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['marital_n'] = le_marital.fit_transform(inputs['marital'])
inputs['education_n'] = le_education.fit_transform(inputs['education'])
inputs['default_n'] = le_default.fit_transform(inputs['default'])
inputs['housing_n'] = le_housing.fit_transform(inputs['housing'])
inputs['loan_n'] = le_loan.fit_transform(inputs['loan'])
inputs['contact_n'] = le_contact.fit_transform(inputs['contact'])
inputs['month_n'] = le_month.fit_transform(inputs['month'])
inputs['poutcome_n'] = le_poutcome.fit_transform(inputs['poutcome'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['job','marital','education','default','housing','loan','contact','month','poutcome'],axis='columns')
```

```python
inputs_n
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.2, random_state=1)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```
